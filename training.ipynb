```python
import matplotlib.pyplot as pylot
import numpy
```

### Visualisation

```python
pylot.plot()
```

```python
x = [i * 10 for i in range(1, 6)]
y = numpy.array(x) ** 2
print(f"| {y}")
print(f"_ {x}")

pylot.plot(x, y, marker="o", color="#333", markeredgecolor="white")
pylot.ylabel("Vertical ( y )")
pylot.xlabel("Horizontal ( x )")
pylot.show()
```

```python
pylot.subplot(121)
pylot.plot(x, y, color="red", marker="o", markeredgecolor="white")

pylot.subplot(122)
pylot.plot(x, y, color="royalblue", marker="o", markeredgecolor="white")

pylot.tight_layout()
pylot.show()
```

### Object Oriented Programming

```python
x = numpy.arange(1, 20, 0.5)
y = x ** 2
print(x)
print("--------------------------")
print(y)
```

```python
fig = pylot.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_ylabel("Vertical (y)")
axes.set_xlabel("Horizontal (x)")
axes.set_title("Graphics")
axes
```

```python
fig = pylot.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # chart movement
axes2 = fig.add_axes([0.25, 0.50, 0.25, 0.25]) # chart movement

axes1.plot(x, y, color="red")
axes1.set_xlabel("Large (x)")
axes1.set_ylabel("Large (y)")
axes1.set_title("Large Graphic")

axes2.plot(y, x, color="royalblue")
axes2.set_xlabel("Small (x)")
axes2.set_ylabel("Small (y)")
axes2.set_title("Small Graphic")

pylot.show()
```

```python
fig, axes = pylot.subplots(1, 2)
print(axes)

axes[0].plot(x, y, color="royalblue")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")
axes[0].set_title("Title 1")

axes[1].plot(y, x, color="mediumseagreen")
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")
axes[1].set_title("Title 2")

fig.tight_layout()
```

### Graphics Management

```python
fig = pylot.figure(figsize=(10, 4), dpi=100)
axes = fig.add_axes([0, 0, .5, 1])
axes.plot(x, y)
pylot.show()
```

```python
fig, axes = pylot.subplots(2, 1, figsize=(4, 4), dpi=150)
axes[0].plot(x, y)
axes[0].set_title("Graphic 1")

axes[1].plot(x, y)
axes[1].set_title("Graphic 2")

pylot.tight_layout()
```

```python
fig = pylot.figure(facecolor="gainsboro", dpi=100)
print(x)
print()
print(y)
ax = fig.add_axes([0, 0, 1.5, 1])
ax.plot(x, y, label="Level 2")
ax.plot(x, x ** 3, label="Level 3")
ax.legend(loc=(.02, .85))
ax.grid(True)
pylot.show()
```

### Seaborn

```python
import pandas
import seaborn
```

```python
titanic = pandas.read_csv('./data/kapal_titanic.csv')
titanic.head()
```

```python
seaborn.set()

# Distribution Plot
seaborn.distplot(titanic["age"], kde=True, bins=100)
```

```python
seaborn.jointplot(x="age", y="fare", data=titanic)
```

```python
import scipy.stats as stat
```

```python
seaborn.jointplot('age', 'fare', stat_func=stat.pearsonr, data=titanic)
```
