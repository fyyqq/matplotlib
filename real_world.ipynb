```python
import pandas
import numpy
import matplotlib.pyplot as pylot
```

Gas Price Dataset

```python
dataset_gas = pandas.read_csv('./data/gas_prices.csv')
dataset_gas.head()
```

```python
pylot.plot(dataset_gas["Year"], dataset_gas["USA"], marker="o", markersize=4, color="royalblue", label="USA")
pylot.plot(dataset_gas["Year"], dataset_gas["Canada"], marker="o", markersize=4, color="red", label="Canada")

pylot.title("USA vs Canada Gas Prices")
pylot.ylabel("Dollar/Gallon")
pylot.legend()

data = int(numpy.ceil(dataset_gas["Year"].count() / 8)) # display 8 years
pylot.xticks(dataset_gas["Year"][::data].tolist() + [2011])

pylot.show()
```

```python
data = numpy.ceil(dataset_gas["Year"].count() / 8)
dataset_gas["Year"][::int(data)].tolist()
```

```python
countries = dataset_gas[["USA", "Canada", "South Korea", "Australia"]].columns
```

```python
dataToManage = {
    "country" : list(countries),
    "color": ["royalblue", "red", "green", "gold"]
}

index = 0
while index < len(dataToManage["country"]) and index < len(dataToManage["color"]):
    pylot.plot(dataset_gas["Year"], dataset_gas[dataToManage["country"][index]], marker="o", color=f"{dataToManage['color'][index]}", markersize="4", label=f"{dataToManage['country'][index]}")
    index = index + 1

pylot.title("Gas Prices Over Time (USD)", fontdict={ "fontsize": "10" })
pylot.xticks(dataset_gas["Year"][::3].tolist() + [2011])
pylot.xlabel("Year")
pylot.ylabel("US Dollars")

pylot.legend()

pylot.show()
```

FIFA Dataset

```python
fifa_dataset = pandas.read_csv('./data/fifa_data.csv')
fifa_dataset.head()
```

Histogram

```python
bins = numpy.array([i * 10 for i in range(11)])

pylot.hist(fifa_dataset["Overall"], bins=bins, color='royalblue')

pylot.xticks(bins)

pylot.xlabel('Numbers Of Players')
pylot.ylabel('Skill Level')

pylot.title("Distribution of Player Skills in FIFA 2018")

pylot.show()
```

Pie Chart

```python
left = fifa_dataset.loc[fifa_dataset["Preferred Foot"] == "Left"].count()[0]
right = fifa_dataset.loc[fifa_dataset["Preferred Foot"] == "Right"].count()[0]

labels = ["Left", "Right"]
colors = ["royalblue", "red"]

pylot.pie([left, right], labels=labels, colors=colors, autopct='%.2f %%')
pylot.title("Foot Preference of FIFA Players")

pylot.show()
```

```python
fifa_dataset["Weight"] = pandas.to_numeric([x.strip('lbs') if type(x) == str else x for x in fifa_dataset["Weight"]])

pylot.style.use('ggplot')

light = fifa_dataset.loc[fifa_dataset["Weight"] < 125].count()[0] # 41
light_medium = fifa_dataset.loc[(fifa_dataset["Weight"] >= 125) & (fifa_dataset["Weight"] < 150)].count()[0] # 2290
medium = fifa_dataset.loc[(fifa_dataset["Weight"] >= 150) & (fifa_dataset["Weight"] < 175)].count()[0] # 10876
medium_heavy = fifa_dataset.loc[(fifa_dataset["Weight"] >= 175) & (fifa_dataset["Weight"] < 200)].count()[0] # 4583
heavy = fifa_dataset.loc[(fifa_dataset["Weight"] >= 200)].count()[0] # 369

pylot.title('Weight Distribution of FIFA Players (in LBS)')


pylot.pie([light, light_medium, medium, medium_heavy, heavy], labels=["light", "light_medium", "medium", "medium_heavy", "heavy"], autopct='%.2f %%', pctdistance=.6, explode=([.1 for _ in range(5)]))
pylot.show()
```

```python
fcb = fifa_dataset.loc[fifa_dataset["Club"] == "FC Barcelona"]["Overall"]
rm = fifa_dataset.loc[fifa_dataset["Club"] == "Real Madrid"]["Overall"]
juv = fifa_dataset.loc[fifa_dataset["Club"] == "Juventus"]["Overall"]

pylot.style.use('default')
pylot.figure(figsize=(5, 8))

pylot.title("Professional Soccer Team Comparison")
pylot.boxplot([fcb, rm, juv], labels=["Barcelona", "Real Madrid", "Juventus"])
pylot.ylabel("FIFA Overall Ratings")
pylot.show()
```
