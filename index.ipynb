```python
import pandas
import numpy
import matplotlib.pyplot as pylot
```

Basic Graph

```python
x = numpy.array([0, 6])
y = numpy.array([0, 250])

pylot.plot(x, y, 'o') # create graph # x = bottom | y = left
pylot.show()
```

```python
x = numpy.array([1, 4, 2, 7])
# y = numpy.array([3, 8, 1, 10])

# pylot.plot(x, y)
pylot.plot(x)
pylot.show()
```

```python
list = numpy.array([2, 4, 3, 12, 3, 5])

pylot.plot(list, marker='o')
pylot.show()
```

```python
list = numpy.array([2, 7, 1, 3, 0, 5, 2])

# pylot.plot(list, linestyle="dotted")
# pylot.plot(list, linestyle="dashed")
# pylot.plot(list, ls="dashed", color="black")
pylot.plot(list, ls="dashed", color="grey", linewidth="5")
pylot.show()
```

```python
x = numpy.array([2, 1, 6, 3])
y = numpy.array([3, 5, 4, 9])

pylot.plot(x, color="red")
pylot.plot(y, color="royalblue")
pylot.show()
```

```python
x = numpy.array([2, 1, 6, 3])
y = numpy.array([3, 5, 4, 9])

style = {
    "color" : "#333",
    "size" : "10"
}

pylot.plot(x, y, marker="o", color="red")
pylot.title("Normal Graph", fontdict=style, loc="center")
pylot.xlabel("horizontal line")
pylot.ylabel("vertical line")
pylot.grid(axis='x')
pylot.grid(axis='y')
pylot.show()
```

```python
pylot.subplot(2, 1, 1)
pylot.plot([0, 1], [0, 1])
pylot.title('Subplot 1')

pylot.subplot(2, 1, 1)
pylot.plot([0, 1], [1, 0])
pylot.title('Subplot 2')
pylot.show()
```

```python
x = numpy.array([2, 4, 1, 5])
y = numpy.array([5, 3, 2, 2])
pylot.scatter(x, y, color="red")

x = numpy.array([5, 3, 2, 2])
y = numpy.array([2, 4, 1, 5])
pylot.scatter(x, y, color="black")

pylot.colorbar()

pylot.show()
```

```python
x = numpy.array(["A", "B", "C", "D"])
y = numpy.array([3, 8, 1, 10])

pylot.bar(x, y, color="red")
pylot.show()
```

```python
x = numpy.array(["A", "B", "C", "D"])
y = numpy.array([3, 8, 1, 10])

pylot.barh(x, y, color="royalblue")
pylot.show()
```

```python
x = numpy.random.normal(120, 10, 250)
pylot.hist(x)
pylot.show()
```

```python
x = numpy.random.randn(1000)

pylot.hist(x, bins=15, color='red', edgecolor='white')

pylot.show()
```

```python
num = 100 / 5

data = numpy.array([30, 50, 10, 10])
label = ["Iphone", "Earbuds", "Macbook", "Vision"]
positioning = [0, 0, 0, .1]

pylot.pie(data, labels=label, startangle=90, explode=positioning)
pylot.legend(title="Product Sales")
pylot.show()
```

PLOT CHART

```python
x = numpy.array([1, 2, 3, 4])
y = numpy.array([5, 6, 7, 8])

pylot.figure(figsize=(10, 5), dpi=100)

pylot.plot(x, y, label='line style', color='purple', linewidth=2, marker='o', markersize='8', markeredgecolor="cyan", linestyle="dashed")

x2 = numpy.arange(0, 4.5, 0.5)
pylot.plot(x2[:6], x2[:6]**2, color="red", label="X")
pylot.plot(x2[5:], x2[5:]**2, color="green", label="X")

pylot.xlabel('X Axis')
pylot.ylabel('Y Axis')

pylot.title("Bitcoin Chart")

# pylot.xticks(x)
# pylot.yticks(y)

pylot.legend()

pylot.savefig('./images/mygraph.png', dpi=100)

pylot.show()
```

BAR CHART

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = pylot.bar(labels, values, color="royalblue")

bars[0].set_hatch('/')
bars[1].set_hatch('*')
bars[2].set_hatch('o')

pylot.figure(figsize=(6, 4))

pylot.show()
```
