```python
def barplot():
    data1 = [25, 85, 75, 40, 60]
    data2 = [40, 35, 20, 55, 10]
    category = ['A', 'B', 'C', 'D', 'E']

    x = numpy.arange(len(category))

    width = 0.35

    pylot.bar(x - width / 2, data1, width, label='Data 1')
    pylot.bar(x + width / 2, data2, width, label='Data 2')

    pylot.grid(linestyle='--', linewidth=1, axis='y', alpha=0.5, color='black')

    pylot.xticks(x, category)
    pylot.xlabel("Category")
    pylot.ylabel("Total")

    pylot.legend()
    pylot.tight_layout()
    pylot.show()

barplot()
```

```python
def C2F(celcius=0):
    return (celcius * 1.8) + 32

def convertion_axes(ax1):
    y1, y2 = ax1.get_ylim()
    ax2.set_ylim(C2F(y1), C2F(y2))
    ax2.figure.canvas.draw()

fig, ax1 = pylot.subplots()
ax2 = ax1.twinx()

ax1.callbacks.connect('ylim_changed', convertion_axes)
ax1.plot(x, 'go-', markersize=3)

print(x)
print()
print(ax1.get_ylim())
print(ax1.get_xlim())
print()
print(ax2.get_ylim())
print(ax2.get_xlim())

ax1.set_xlabel('Day')
ax1.set_ylabel('Celcius')
ax2.set_ylabel('Fahrenheit')

fig.suptitle("Temprature wind")
pylot.show()
```

Work with Datasets

```python
dataset = pandas.read_csv('./data/2020.csv')
dataset.rename(columns={
    "Country name": "Country",
    "Perceptions of corruption": "Corruption",
    "Explained by: Social support": "Social Support",
    "Explained by: Freedom to make life choices": "Freedom"
}, inplace=True)

dataset.head()
```

```python
dataset[['Country', 'Corruption', 'Freedom', 'Generosity', 'Social Support']]
```

```python
dataset.iloc[:3].plot(kind='bar', x='Country', y=['Corruption', 'Freedom', 'Generosity', 'Social Support'])

pylot.title("Bar Plot")
pylot.ylabel("Total")

pylot.show()
```

```python
dataset[:5].plot(kind='line', x='Country', y=['Corruption', 'Freedom', 'Generosity', 'Social Support'], marker="o", markersize=4)

pylot.title("Line Plot")
pylot.ylabel("Value")

pylot.show()
```

```python
dataset[:5].plot(kind='box', x='Country', y=['Corruption', 'Freedom', 'Generosity', 'Social Support'])

pylot.title("Box Plot")
pylot.ylabel("Value")

pylot.show()
```

```python
dataset[:5].plot(kind='scatter', x='Country', y='Freedom')

pylot.title("Scatter Plot")
pylot.ylabel("Value")

pylot.show()
```

```python
dataset.plot(kind='hist', x='Country', y='Corruption', bins=20, color='royalblue')

pylot.title("Histogram Plot")
pylot.ylabel("Value")

pylot.show()
```

```python
dataset.plot(kind='hist', x='Country', y=['Corruption', 'Freedom', 'Generosity', 'Social Support'], subplots=True, layout=(2, 2))

pylot.title("Subplot Plot")

pylot.tight_layout()
pylot.show()
```
